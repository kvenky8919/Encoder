### OneHotEncoder

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder()
```

```python
ohe
```

```python
ohe.fit(dfrev[["gender"]])
```

```python
ohe.transform(dfrev[["gender"]])
```

```python
ary = ohe.transform(dfrev[["gender"]]).toarray()
```

```python
ary
```

```python
ohe.categories_
```

```python
df1 = pd.DataFrame(ary, dtype = "int",columns = ['Female', 'Male'] )
```

```python
df1
```

```python
df = pd.concat([dfrev,df1], axis =1)
```

```python
df
```
